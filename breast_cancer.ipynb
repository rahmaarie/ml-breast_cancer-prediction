# **1. Install Library**

# **2. Read Dataset**

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer() #Memuat dataset dari Scikit-Learn
x = cancer.data  # Fitur
y = cancer.target  # Label

df_x = pd.DataFrame(x, columns=cancer.feature_names)
df_y = pd.Series(y, name='target')

print("Jumlah fitur sebelum dihapus:", df_x.shape[1])  # Seharusnya 30

# Hapus 1 fitur yang tidak digunakan agar hanya 29 fitur
df_x = df_x.drop(columns=["mean perimeter"])  # Contoh, hapus salah satu fitur

print("Jumlah fitur setelah dihapus:", df_x.shape[1])  # Harus 29
```

#**3. Split Data**

```python
# Membagi dataset menjadi data latih dan data uji(test_size) dengan proporsi 70:30.
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.2, random_state=42)
```

# **4. Train the Model**

```python
from sklearn.ensemble import RandomForestClassifier

# Membangun dan melatih model Random Forest dengan 100 pohon keputusan.
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(x_train, y_train)
```

# **5. Predict & Evaluate**

```python
from sklearn.metrics import accuracy_score, confusion_matrix

# Memprediksi label pada data uji dan mengevaluasi performa model menggunakan akurasi.
y_pred = model.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)

print("Laporan Klasifikasi:")
print(f"Akurasi: {accuracy * 100:.2f}%")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=cancer.target_names, yticklabels=cancer.target_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import accuracy_score, classification_report

print(classification_report(y_test, y_pred,target_names=cancer.target_names))
```

Mengevaluasi performa model dalam hal true positive rate (TPR) dan false positive rate (FPR). ROC curve membantu memahami trade-off antara TPR dan FPR, serta memberikan gambaran tentang performa model dalam berbagai threshold.
